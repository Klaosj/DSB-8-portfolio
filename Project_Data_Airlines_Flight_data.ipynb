Dataset: Airlines Flights Dataset for different cities

```python
import numpy as np
import plotly.express as px
import tabulate as tb
import matplotlib.pyplot as plt
import pandas as pd
import plotly.graph_objects as go
import heapq
import kagglehub
from kagglehub import KaggleDatasetAdapter

file_path = "airlines_flights_data.csv"
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "rohitgrewal/airlines-flights-data",
  file_path,)
```

Check infomation and null data

```python
display(print(df.isnull().sum()))
display(df.head())
```

Q.1 What are the airlines in the dataset, accompanied by their frequencies?

```python
df1 = df.groupby('airline').size().reset_index(name='frequency')
df1 = df1.sort_values(by='frequency', ascending=False)
fig = px.bar(df1, x='airline', y='frequency')
fig.show()

display(df1)
```

Q.2. Show Bar Graphs representing the Departure Time & Arrival Time.

```python
df2 = df.groupby('departure_time').size().reset_index(name = 'count_time')
df3 = df.groupby('arrival_time').size().reset_index(name = 'count_time')

df2 = df2.sort_values(by= 'count_time', ascending= False)

fig = go.Figure(data=[
    go.Bar(name='departure_time', x=df2['departure_time'], y=df2['count_time']),
    go.Bar(name='arrival_time', x=df3['arrival_time'], y=df3['count_time'])
])
# Change the bar mode
fig.update_layout(barmode='group')
fig.show()

display(df2)
display(df3)
```

Q.3. Show Bar Graphs representing the Source City & Destination City.

```python
df4 = df.groupby('source_city').size().reset_index(name = 'count_city')
df5 = df.groupby('destination_city').size().reset_index(name = 'count_city')

fig = go.Figure(data=[
    go.Bar(name='source_city', x=df4['source_city'], y=df4['count_city']),
    go.Bar(name='destination_city', x=df5['destination_city'], y=df5['count_city'])
])
# Change the bar mode
fig.update_layout(barmode='group')
fig.show()

display(df4)
display(df5)
```

Q.4 Does price varies with airlines ?

```python
df7 = round(df.groupby(['airline','class'])['price'].mean().reset_index(name = 'price'), 2)

Economy_df = df7[df7['class'] == 'Economy']
Business_df = df7[df7['class'] == 'Business']

fig = px.bar(Economy_df, x='airline', y='price')
fig.update_layout(title='Airline Economy class ')
fig.show()

fig = px.bar(Business_df, x='airline', y='price')
fig.update_layout(title='Airline Business class ')
fig.show()


display(Economy_df)
display(Business_df)
```

Q.5. Does ticket price change based on the departure time and arrival time?

```python
Economy_df = df[df['class'] == 'Economy']
df8 = round(Economy_df.groupby('departure_time')['price'].mean().reset_index(name = 'price'), 2)
df9 = round(Economy_df.groupby('arrival_time')['price'].mean().reset_index(name = 'price'), 2)

fig = go.Figure(data=[
    go.Bar(name='departure_time_Economy', x=df8['departure_time'], y=df8['price']),
    go.Bar(name='Arrival_time_Economy', x=df9['arrival_time'], y=df9['price'])
])
# Change the bar mode
fig.update_layout(title='Different Price base on departure and arrival time ')
fig.update_layout(barmode='group')
fig.show()


display(df8)
display(df9)
```

Q.6. How the price changes with change in Source and Destination?

```python
# Filter for Economy class
Economy_df = df[df['class'] == 'Economy']

df10 = round(Economy_df.groupby('source_city')['price'].mean().reset_index(name = 'price'), 2)
df11 = round(Economy_df.groupby('destination_city')['price'].mean().reset_index(name = 'price'), 2)

fig = go.Figure(data=[
    go.Bar(name='source_city', x=df10['source_city'], y=df10['price']),
    go.Bar(name='destination_city', x=df11['destination_city'], y=df11['price'])
])
fig.update_layout(title='Different economy price base on source city and destination city')
fig.update_layout(barmode='group')
fig.show()


# Filter for Business class
Business_df = df[df['class'] == 'Business']

df12 = round(Business_df.groupby('source_city')['price'].mean().reset_index(name = 'price'), 2)
df13 = round(Business_df.groupby('destination_city')['price'].mean().reset_index(name = 'price'), 2)

fig = go.Figure(data=[
    go.Bar(name='source_city', x=df12['source_city'], y=df12['price']),
    go.Bar(name='destination_city', x=df13['destination_city'], y=df13['price'])
])
fig.update_layout(title='Different business price base on source city and destination city')
fig.update_layout(barmode='group')
fig.show()

# Number table
print('Different economy price base on source city and destination city')
display(df10)
display(df11)

print('Different Business price base on source city and destination city')
display(df12)
display(df13)
```

Q.7. How is the price affected when tickets are bought in just 1 or 2 days before departure?

```python
Economy_df = df[df['class'] == 'Economy']
df14 = round(Economy_df.groupby(['days_left'])['price'].mean().reset_index(name = 'price'), 2)

fig = px.area(df14, x='days_left', y='price', color_discrete_sequence=['orange'])
fig.update_layout(title='Different price base on days left in Economy Class')
fig.show()

Business_df = df[df['class'] == 'Business']
df15 = round(Business_df.groupby(['days_left'])['price'].mean().reset_index(name = 'price'), 2)

fig = px.area(df15, x='days_left', y='price', color_discrete_sequence=['purple'])
fig.update_layout(title='Different price base on days left in Business Class')
fig.show()
```
